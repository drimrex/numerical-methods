### 2 практическая

```python
from math import *
def f(x):
  n = (10 * cos(x)) - (0.1 * (x ** 2))
  return n
eps = 1
a = 7
b = 8
while eps > 0.001:
  c = (a + b)/2
  if f(a) * f(c) < 0:
    b = c
  else:
    a = c
  eps = (b - a)/2
answer = (a + b)/2
print('Ответ: ',round(answer, 2))
```
